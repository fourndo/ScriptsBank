```python
import numpy as np
import requests
from requests.exceptions import RequestException
from contextlib import closing
from bs4 import BeautifulSoup
import pandas as pd
import json
import matplotlib.pyplot as plt
def simple_get(url):
    """
    Attempts to get the content at `url` by making an HTTP GET request.
    If the content-type of response is some kind of HTML/XML, return the
    text content, otherwise return None.
    """
    try:
        with closing(get(url, stream=True)) as resp:
            if is_good_response(resp):
                return resp.content
            else:
                return None

    except RequestException as e:
        log_error('Error during requests to {0} : {1}'.format(url, str(e)))
        return None


def is_good_response(resp):
    """
    Returns True if the response seems to be HTML, False otherwise.
    """
    content_type = resp.headers['Content-Type'].lower()
    return (resp.status_code == 200 
            and content_type is not None 
            and content_type.find('html') > -1)


def log_error(e):
    """
    It is always a good idea to log errors. 
    This function just prints them, but you can
    make it do anything.
    """
    print(e)



class HTMLTableParser:

    def parse_url(self, url):
        response = requests.get(url)
        soup = BeautifulSoup(response.text,'html.parser')
#         print(soup)

        collector = []
        for table in soup.find_all('table'):
            if table.has_attr('id'):
                if table['id'] == "games":
                    collector += [(table['id'],self.parse_html_table(table)) ]
                
        return collector 

    def parse_html_table(self, table):
        n_columns = 0
        n_rows=0
        column_names = []

        # Find number of rows and columns
        # we also find the column titles if we can
        for row in table.find_all('tr'):

            # Determine the number of rows in the table
            td_tags = row.find_all('td')
            if len(td_tags) > 0:
                n_rows+=1
                if n_columns == 0:
                    # Set the number of columns for our table
                    n_columns = len(td_tags)

            # Handle column names if we find them
            th_tags = row.find_all('th') 
            if len(th_tags) > 0 and len(column_names) == 0:
                for th in th_tags:
                    column_names.append(th.get_text())

                    
        column_names = column_names[1:]
        column_names[1] = 'Home'
        column_names[5] = 'W_L'
        column_names[6] = 'overtimes'
        
        # Safeguard on Column Titles
        if len(column_names) > 0 and len(column_names) != n_columns:
            raise Exception("Column titles do not match the number of columns")

        columns = column_names if len(column_names) > 0 else range(0,n_columns)
        df = pd.DataFrame(columns = columns,
                          index= range(0,n_rows))
        row_marker = 0
        for row in table.find_all('tr'):
            column_marker = 0
            columns = row.find_all('td')

            if len(columns) > 0:
                for column in columns:
                    df.iloc[row_marker,column_marker] = column.get_text()
                    column_marker += 1
                if len(columns) > 0:
                    row_marker += 1

#         # Convert to float if possible
#         for col in df:
#             try:
#                 df[col] = df[col].astype(float)
#             except ValueError:
#                 pass

        return df

```

```python
with open('./nhlTeamCallSign.json', 'r') as fp:
    teams = json.load(fp)
    
```

```python
games2018 = {}

for team in list(teams.keys()):
    url = 'https://www.hockey-reference.com/teams/'+ team + '/2018_games.html'

    hp = HTMLTableParser()
    table = hp.parse_url(url)[0][1] # Grabbing the table from the tuple
    
    games2018[team] = table
```

Previous game win, next game home , pct win/lose

```python
# Condition
# Win -> homegame win

winHome = []
winVisi = []
for team in list(games2018.keys()):
    subData = games2018[team].W_L.values[1:]
    
    indCondition = (games2018[team].W_L.values[:-1] == "W") * (games2018[team].Home.values[1:]=='@')
    winVisi += [np.sum(subData[indCondition]=='W')/np.sum(indCondition)]
    
    indCondition = (games2018[team].W_L.values[:-1] == "W") * (games2018[team].Home.values[1:]=='')
    winHome += [np.sum(subData[indCondition]=='W')/np.sum(indCondition)]
```

```python
fig = plt.figure(figsize=(15,5))
ax = plt.subplot(2,1,1)
N = len(list(games2018.keys()))
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, winVisi, width)
# p2 = plt.bar(ind, winHome, width)
plt.plot([0,N], [0.5, 0.5], "k--")
plt.plot([0,N], [np.mean(winVisi), np.mean(winVisi)], "r--")
plt.ylabel('$\%$')
plt.title('$\%$ Win + Win visiting game')
plt.xticks(ind, list(games2018.keys()))
ax.set_ylim([0,1])
# plt.yticks(np.arange(0, 81, 10))
# plt.legend((p2[0], p1[0]), ('Win Visitor', 'Win Home'))

ax = plt.subplot(2,1,2)
p1 = plt.bar(ind, winHome, width)
# p2 = plt.bar(ind, winHome, width)
plt.plot([0,N], [0.5, 0.5], "k--")
plt.plot([0,N], [np.mean(winHome), np.mean(winHome)], "r--")
plt.ylabel('$\%$')
plt.title('$\%$ Win + Win Home game')
plt.xticks(ind, list(games2018.keys()))
ax.set_ylim([0,1])
```
