```python
import numpy as np
import scipy as sp
import unittest
from SimPEG import Mesh, Maps, Models, Utils
import inspect
import matplotlib.pyplot as plt
a = np.array([1, 1, 1])
b = np.array([1, 2])

mesh2 = Mesh.TensorMesh([a, b], x0=np.array([3, 5]))
mesh3 = Mesh.TensorMesh([a, b, [3, 4]], x0=np.array([3, 5, 2]))
mesh22 = Mesh.TensorMesh([b, a], x0=np.array([3, 5]))
```

```python
# Create sub meshes offseted
mesh2b = Mesh.TensorMesh([np.r_[mesh2.hy,mesh2.hy], mesh2.hy], x0 = mesh2.x0)
mesh2b._x0[0] = mesh2b.x0[0]-2
actv2 = np.ones(mesh2.nC, dtype='bool')
actv2b = np.ones(mesh2b.nC, dtype='bool')
# Create model
m = np.ones(mesh2.nC)
m[4] = 2

M = Maps.Tile((mesh2, actv2), (mesh2b, actv2b))
mout = M.Paverage*m
```

```python
mesh2.plotImage(m)
plt.show()

mesh2b.plotImage(mout, grid='on')
plt.show()
```

```python
mout[mout>0].max()
```

```python
tol = 1e-7

# Check that max and min model values are preserved
np.linalg.norm(mout[mout>0].min()-m.min()) < tol
np.linalg.norm(mout[mout>0].max()-m.max()) < tol
```

```python
octree = Mesh.TreeMesh([np.ones(4),np.ones(4),np.ones(4)], x0='CCC')
# octree._x0=np.asarray([0,0,0])
octree.refine(6)

def refine(cell):
    xyz = cell.center
    for i in range(3):
        if (xyz[0]**2. + xyz[1]**2.)**0.5 < 1:
            return 6-i
    return 0

octree.refine(refine)
# octree.writeUBC('OctreeTest.msh')
octree.cell
```

```python
M = Mesh.TreeMesh([32, 32, 32])
M.refine(2)
def refine(cell):
    xyz = cell.center
    if np.logical_and(xyz[0]>0.2, xyz[0]<0.8) and np.logical_and(xyz[1]>0.2, xyz[1]<0.8) and np.logical_and(xyz[2]>0.2, xyz[2]<0.8):
        return 4
    return 0
M.refine(refine)

ax = plt.subplot(projection='3d')
M.plotGrid(ax=ax)
ax.view_init(90, 90)
plt.show()
```
